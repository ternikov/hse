```python
# All keys
gradebook.keys()
```

```python
# All values
gradebook.values()
```

```python
# View all values
for k in gradebook:
    print(k)
```

```python
# Pairs of key-value (ver. 1)
for k in gradebook:
    print(k, '-', gradebook[k])
```

```python
# Pairs of key-value (ver. 2)
for k, v in gradebook.items():
    print(k, "-", v)
```

```python
students = ["A", "B", "C", "D"]
grades = [5, 4, 2, 3]
new_gradebook = dict(zip(students, grades))
new_gradebook
```
